```python

team_schema = StructType([
    StructField("team_sk", IntegerType(), True),
    StructField("team_id", IntegerType(), True),
    StructField("team_name", StringType(), True),
])
```

```python
team_df = spark.read.schema(team_schema ).format("csv").option("header", "true").load("/content/drive/MyDrive/ipl-data-till-2017/Team.csv")
```

```python
from pyspark.sql.functions import col, count, desc, avg, sum as spark_sum

# Total matches per season
match_df.groupBy("season_year").count().orderBy("season_year").show()

# Most wins by team
match_df.groupBy("match_winner").count().orderBy(desc("count")).show()

# Toss winner vs match winner correlation
match_df.withColumn(
    "toss_match_same", col("toss_winner") == col("match_winner")
).groupBy("toss_match_same").count().show()

# Wins by win type (runs vs wickets)
match_df.groupBy("win_type").count().show()

# Most Man of the Match awards
match_df.groupBy("manofmach").count().orderBy(desc("count")).show(10)
```

```python
# Top 10 run scorers overall
ball_by_ball_df.groupBy("striker") \
    .agg(spark_sum("runs_scored").alias("total_runs")) \
    .orderBy(desc("total_runs")).show(10)

# Most boundaries (4s and 6s)
ball_by_ball_df.filter(col("runs_scored").isin(4, 6)) \
    .groupBy("striker", "runs_scored").count() \
    .orderBy(desc("count")).show(10)

# Average runs per match per team
ball_by_ball_df.groupBy("match_id", "team_batting") \
    .agg(spark_sum("runs_scored").alias("runs")) \
    .groupBy("team_batting") \
    .agg(avg("runs").alias("avg_runs_per_match")) \
    .orderBy(desc("avg_runs_per_match")).show()

# Total extras per team
ball_by_ball_df.groupBy("team_batting") \
    .agg(spark_sum("extra_runs").alias("total_extras")) \
    .orderBy(desc("total_extras")).show()
```

```python
# Enrich ball data with match info
ball_match_df = ball_by_ball_df.join(match_df, on="match_id", how="inner")

# Player name with total runs (join Player)
ball_by_ball_df.join(player_df, ball_by_ball_df.striker == player_df.player_id, "inner") \
    .groupBy("player_name") \
    .agg(spark_sum("runs_scored").alias("total_runs")) \
    .orderBy(desc("total_runs")).show(10)

# Man of the match with player details
match_df.join(player_df, match_df.manofmach == player_df.player_name, "left") \
    .select(player_df["player_name"], player_df["country_name"].alias("player_country")) \
    .groupBy("player_name", "player_country") \
    .count().orderBy(desc("count")).show(10)
```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import rank, dense_rank

# Top scorer per season
season_runs = ball_by_ball_df.join(match_df, "match_id") \
    .groupBy("season_year", "striker") \
    .agg(spark_sum("runs_scored").alias("total_runs"))

window = Window.partitionBy("season_year").orderBy(desc("total_runs"))

season_runs.withColumn("rank", rank().over(window)) \
    .filter(col("rank") == 1).show()

# Running total of runs per match per team
run_window = Window.partitionBy("match_id", "team_batting").orderBy("over_id", "ball_id")
from pyspark.sql.functions import sum as spark_sum
ball_by_ball_df.withColumn("running_total", spark_sum("runs_scored").over(run_window)).show(20)
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
wins_pd = match_df.groupBy("match_winner").count() \
    .orderBy("count", ascending=False).limit(10).toPandas()

plt.figure(figsize=(12, 5))
sns.barplot(data=wins_pd, x="match_winner", y="count", palette="YlOrRd_r")
plt.title("Top 10 Teams by Total Wins", fontsize=15, fontweight="bold")
plt.xlabel("Team"); plt.ylabel("Wins")
plt.xticks(rotation=30, ha="right")
plt.tight_layout()
plt.show()
```

```python
season_runs = ball_by_ball_df.join(match_df.select("match_id","season_year"), "match_id") \
    .groupBy("season_year") \
    .agg(spark_sum("runs_scored").alias("total_runs")) \
    .orderBy("season_year").toPandas()

plt.figure(figsize=(12, 5))
plt.plot(season_runs["season_year"], season_runs["total_runs"],
         marker="o", linewidth=2.5, color="#e84545")
plt.fill_between(season_runs["season_year"], season_runs["total_runs"], alpha=0.15, color="#e84545")
plt.title("Total Runs Scored Per Season", fontsize=15, fontweight="bold")
plt.xlabel("Season"); plt.ylabel("Total Runs")
plt.xticks(season_runs["season_year"], rotation=45)
plt.tight_layout()
plt.show()
```

```python
top_batsmen = ball_by_ball_df.join(player_df, ball_by_ball_df.striker == player_df.player_id) \
    .groupBy("player_name") \
    .agg(spark_sum("runs_scored").alias("total_runs")) \
    .orderBy("total_runs", ascending=False).limit(10).toPandas()

plt.figure(figsize=(10, 6))
sns.barplot(data=top_batsmen, y="player_name", x="total_runs", palette="Blues_r")
plt.title("Top 10 Run Scorers (All Time)", fontsize=15, fontweight="bold")
plt.xlabel("Total Runs"); plt.ylabel("Player")
plt.tight_layout()
plt.show()
```

```python
over_runs = ball_by_ball_df.groupBy("match_id", "innings_no", "over_id") \
    .agg(spark_sum("runs_scored").alias("over_runs")).toPandas()

plt.figure(figsize=(16, 5))
sns.boxplot(data=over_runs[over_runs["over_id"] <= 20],
            x="over_id", y="over_runs", palette="coolwarm")
plt.title("Run Distribution Per Over (Overs 1–20)", fontsize=15, fontweight="bold")
plt.xlabel("Over"); plt.ylabel("Runs in Over")
plt.tight_layout()
plt.show()
```

```python
from pyspark.sql.functions import when

margin_pd = match_df.filter(col("win_margin") > 0)\
                    .withColumn("win_type", when(col("win_type") == "run", "runs").otherwise(col("win_type")))\
                    .toPandas()

plt.figure(figsize=(12, 5))
sns.histplot(data=margin_pd, x="win_margin", hue="win_type",
             bins=30, palette={"runs": "#e84545", "wickets": "#3fc1c9"},
             kde=True)
plt.title("Win Margin Distribution by Type", fontsize=15, fontweight="bold")
plt.xlabel("Win Margin"); plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```
